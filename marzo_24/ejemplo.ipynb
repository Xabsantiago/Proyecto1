```python
import scipy.stats as stats 
import numpy as np
import matplotlib.pyplot as plt
```

$X \sim Exp(\lambda = 0.25)$



```python
l=0.25
beta = 1/l
X= stats.expon(scale = beta)
```

```python
x= np.arange(-1,40,0.1)
pdf = X.pdf(x)
plt.plot(x, pdf)
plt.xlabel('X')
plt.ylabel('Probability')
plt.show
```
