```python
import scipy.stats as stats 
import numpy as np
import matplotlib.pyplot as plt
```

$X \sim Exp(\lambda = 0.25)$



```python
l=0.25
beta = 1/l
X= stats.expon(scale = beta)
```

```python
x= np.arange(-1,40,0.1)
pdf = X.pdf(x)
plt.plot(x, pdf)
plt.xlabel('X')
plt.ylabel('Probability')
plt.show
```

```python
plt.figure()
cdf = X.cdf(x)
plt.plot(x, cdf)
plt.xlabel('X')
plt.ylabel('CDF')
plt.show()
```

$$P(4<X<5) = P(X<5) - P(X<4) = P(X \leq 5) - P(X \leq 4)$$
$$P(4<X<5) = F(X=5) - F(X=4)$$ 

```python
P1 = X.cdf(5) - X.cdf(4)
print("P(4 < X < 5) =", P1)
```
